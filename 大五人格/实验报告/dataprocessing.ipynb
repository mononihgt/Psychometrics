```python
selfMean = selfAssessment[['开放性', '尽责性', '外倾性', '宜人性', '神经质']].agg(['mean', 'std']).T
selfMean

peerMean = peerAssessment[['开放性', '尽责性', '外倾性', '宜人性', '神经质']].agg(['mean', 'std']).T

```

```python
from scipy.stats import pearsonr

# 两两计算selfassessment和peerassessment列与列的相关性，如果有5列，则有25个相关，不用计算selfassessment内部的相关性
selfPeerCorr = pd.DataFrame()
for i in range(5):
    for j in range(5):
        col_self = selfMean.index[i]
        col_peer = peerMean.index[j]
        if col_self in selfAssessment.columns and col_peer in peerAssessment.columns:
            corr_value, p_value = pearsonr(selfAssessment[col_self], peerAssessment[col_peer])
            corr_value = f'{round(corr_value, 3)}'
            if p_value < 0.001:
                corr_value = f'{corr_value}**'
            elif p_value < 0.01:
                corr_value = f'{corr_value}**'
            elif p_value < 0.05:
                corr_value = f'{corr_value}*'

            selfPeerCorr.loc[col_peer, col_self] = corr_value
        else:
            print(f'{col_self} or {col_peer} not in the columns')
            selfPeerCorr.loc[col_peer, col_self] = np.nan
print(selfPeerCorr)
```

```python
from scipy.stats import ttest_rel
def cohens_d(data1, data2, relation):
    assert relation in ['paired', 'independent'], 'data relation must be paired or independent'
    if relation == 'paired':
        diff = data1 - data2
        mean_diff = np.mean(diff)
        sd_diff = np.std(diff, ddof=1)
        d = mean_diff / sd_diff
    else:
        mean1, mean2 = np.mean(data1), np.mean(data2)
        var1, var2 = np.var(data1, ddof=1), np.var(data2, ddof=1)
        n1, n2 = len(data1), len(data2)
        pooled_sd = np.sqrt(((n1-1)*var1 + (n2-1)*var2) / (n1 + n2 - 2))
        d = (mean1 - mean2) / pooled_sd
    return d

selfPeerPairT = pd.DataFrame(columns=['t_value', 'p_value', 'cohen_d'])
for i in range(5):
    col = selfMean.index[i]
    t_value, p_value = ttest_rel(selfAssessment[col], peerAssessment[col])
    if p_value < 0.001:
        p_value = f'{p_value}***'
    elif p_value < 0.01:
        p_value = f'{p_value}**'
    elif p_value < 0.05:
        p_value = f'{p_value}*'
    d = cohens_d(selfAssessment[col], peerAssessment[col], 'paired')
    selfPeerPairT.loc[col] = [t_value, p_value, d]
selfPeerPairT
```

```python
from scipy.stats import ttest_ind
bigFive = ['开放性', '尽责性', '外倾性', '宜人性', '神经质']
selfAssessmentMale, selfAssessmentFemale = selfAssessment[selfAssessment['性别'] == '男'].reset_index(drop = True), selfAssessment[selfAssessment['性别'] == '女'].reset_index(drop = True)

genderT = pd.DataFrame(columns=['t_value', 'p_value', 'cohen_d'])
for i in range(len(bigFive)):
    col = bigFive[i]
    t_value, p_value = ttest_ind(selfAssessmentMale[col], selfAssessmentFemale[col])
    if p_value < 0.001:
        p_value = f'{p_value}***'
    elif p_value < 0.01:
        p_value = f'{p_value}**'
    elif p_value < 0.05:
        p_value = f'{p_value}*'
    d = cohens_d(selfAssessmentMale[col], selfAssessmentFemale[col], 'independent')
    genderT.loc[col] = [t_value, p_value, d]

genderT
```
