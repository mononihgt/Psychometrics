```python
import pandas as pd
import numpy as np

rawDf = pd.read_csv('preExp.csv', sep=';')
rawDf.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(rawDf)
scaledDf = scaler.transform(rawDf)
scaledDf = pd.DataFrame(scaledDf, columns=rawDf.columns)
scaledDf.head(3)
# scaledDf.to_csv('preExpScaled.csv', index=False, sep=',')
```

```python
itemAnalysis = pd.read_csv('figure\Item_Analysis.csv', sep=',')
itemAnalysis.rename(columns={'Unnamed: 0':'item'}, inplace=True)
# itemAnalysis.head(3)
itemAnalysis = itemAnalysis[itemAnalysis['RIT']>0.4]
itemAnalysis.head(3)
needItem = itemAnalysis.item.tolist()
preExpScaled = pd.read_csv('preExpScaled.csv', sep=',')
preExpScaled = preExpScaled[needItem]
preExpScaled.to_csv('preExpScaledRevised.csv', index=False)
```
